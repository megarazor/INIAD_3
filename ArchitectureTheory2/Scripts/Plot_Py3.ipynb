```python
import pandas as pd
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
import matplotlib.pyplot as plt
```

```python
py3_seq = pd.read_csv('sequential_py3.2.csv')
py3_thr_sgl = pd.read_csv('threaded_py3.2_singlecore.csv')
py3_thr_mul = pd.read_csv('threaded_py3.2_multicore.csv')
```

```python
x1_raw= py3_seq['Size']
y1= py3_seq['Time']
x2_raw= py3_thr_sgl['Size']
y2= py3_thr_sgl['Time']
x3_raw= py3_thr_mul['Size']
y3= py3_thr_mul['Time']
```

```python
x1= []
x2= []
x3= []
for x in x1_raw:
    converted= str(x//1000000) + "M"
    x1.append(converted)
for x in x2_raw:
    converted= str(x//1000000) + "M"
    x2.append(converted)
for x in x3_raw:
    converted= str(x//1000000) + "M"
    x3.append(converted)
```

```python
plt.plot(x1, y1, label="Sequential")
plt.plot(x2, y2, label="Single-core threaded")
plt.plot(x3, y3, label="4-core threaded")
legend = plt.legend()
plt.title("Performance Test on Python 3.2")
plt.xlabel("Data size")
plt.ylabel("Time (s)")
plt.ylim(0, 70)
# plt.show()
plt.savefig('py3.png')
```
