```python
import pandas as pd
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
import matplotlib.pyplot as plt
```

```python
py2_seq = pd.read_csv('sequential_py2.6.4.csv')
py2_thr_sgl = pd.read_csv('threaded_py2.6.4_singlecore.csv')
py2_thr_mul = pd.read_csv('threaded_py2.6.4_multicore.csv')
py2_thr_mul_itv = pd.read_csv('threaded_py2.6.4_multicore_interval100000.csv')
```

```python
x1_raw= py2_seq['Size']
y1= py2_seq['Time']
x2_raw= py2_thr_sgl['Size']
y2= py2_thr_sgl['Time']
x3_raw= py2_thr_mul['Size']
y3= py2_thr_mul['Time']
x4_raw= py2_thr_mul_itv['Size']
y4= py2_thr_mul_itv['Time']
```

```python
x1= []
x2= []
x3= []
x4= []
for x in x1_raw:
    converted= str(x//1000000) + "M"
    x1.append(converted)
for x in x2_raw:
    converted= str(x//1000000) + "M"
    x2.append(converted)
for x in x3_raw:
    converted= str(x//1000000) + "M"
    x3.append(converted)
for x in x4_raw:
    converted= str(x//1000000) + "M"
    x4.append(converted)
```

```python
x2
```

```python
plt.plot(x1, y1, label="Sequential")
plt.plot(x2, y2, label="Single-core threaded")
plt.plot(x3, y3, label="4-core threaded")
plt.plot(x4, y4, label="4-core threaded, Check Interval = 100,000")
legend = plt.legend()
plt.title("Performance Test on Python 2.6.4")
plt.xlabel("Data size")
plt.ylabel("Time (s)")
plt.ylim(0, 70)
# plt.show()
plt.savefig('py2.png')
```
